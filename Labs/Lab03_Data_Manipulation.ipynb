* When reading in a data set, you may need to change some values.
* Using pandas .loc[], you can SELECT values, but you can also UPDATE values.
* For example, let's select the row with Rex

```python
grades_df.loc[(grades_df['Name']=='Rex')]
```

* You notice that Rex's Final grade is incorrect. It should actually be 83.
* With .loc[] update queries, .loc[<row_indexer>, <col_indexer>]
* You can use a particular value, or a condition to index:

```python
# Here, we are updating Rex's Final grade to be 82.
# The row index is 3, the column index is Final
grades_df.loc[3, 'Final'] = 82
grades_df
```

* Whoops! We set the grade to 82. Change it using a condition instead:

```python
# In the <row_index> position, we use a query to grab the rows with Name == Rex
grades_df.loc[(grades_df['Name']=='Rex'), 'Final'] = 83
grades_df
```

* We need a little bit more information to calculate the grades of the students...
* Below is an additional data set to complete the students grades available for the semester:

```python
more_grades = {
    'Name':['Ariel', 'Brody', 'Debra', 'Rex', 'Jerry', 'Gary', 'Gared', 'Bayley', 'Nathan', 'Heidi'],
    'Attendance':[9, 10, 8, 10, 7, 7, 8, 10, 3, 8],
    'HW':[9, 9, 6, 10, 5, 8, 9, 8, 5, 8]
}
more_grades_df = pd.DataFrame(more_grades)
more_grades_df
```

* Before calculating the grades, we need to JOIN this DataFrame together with the original one.
* Pandas has the .merge() method to perform this.
* NOTE: there is also .join(), but .merge() is more flexible.
* The default is 'left' for .join(), but for .merge() is 'inner'.

```python
new_df = grades_df.merge(more_grades_df, on='Name', how='inner')
new_df
```

* Because merge is more flexible, we can even merge using column names that don't match.

```python
more2 = more_grades_df.copy()
print(more2.columns) #gives us the column names
more2.columns = ['StudName', 'Attendance', 'HW']
print(more2.columns)
more2_merged  = grades_df.merge(more2, left_on='Name', right_on='StudName', how='inner')
more2_merged
```

* We see that using merge, we can select two columns with different names.
* But, do we really need both columns of Names?
* Using the df.drop() command, we can remove the StudName column.
* It's important to note that axis(0) is for rows, and axis(1) is for columns.
* You can easily remember this because an m x n DataFrame is the same as rows x cols...
* If you put this into a tuple (m,n) or (rows,cols), rows is at position 0 and cols is at position 1.

```python
more2_merged.drop('StudName',axis=1, inplace=True)
more2_merged
```

* Now, we need a way to insert a new column based on condition. Let's say the final letter is computer as such:
* A = [100,89] | B+ = [88,83] | B = [82,79] | C+ = [78,73] | C = [72,68] | D = [67,60] | F = < 60

* Another option is to use the .apply() method. This one is interesting because you can take a function(s) and apply them over the entire DataFrame.
* Let's take a look at an example:

```python
new_df[['Midterm','Final','Attendance','HW','Final Average']].apply(np.max)
```

```python
new_df[['Midterm','Final','Attendance','HW','Final Average']].apply(np.min)
```

```python
new_df[['Midterm','Final','Attendance','HW','Final Average']].apply([np.mean, np.median])
```

* But in this case, we want to apply() a function which assigns a letter grade based on the range of the value.
* First, let's define a function which translates a numeric grade into the corresponding letter grade:

```python
def get_letter_grade(average):
    # A = [100,89] | B+ = [88,83] | B = [82,79] | C+ = [78,73] | C = [72,68] | D = [67,60] | F = < 60
    # Assume no average is > 100 or < 0
    if average >= 89:
        return 'A'
    elif average >= 83 and average < 89:
        return 'B+'
    elif average >= 79 and average < 83:
        return 'B'
    elif average >= 73 and average < 79:
        return 'C+'
    elif average >= 68 and average < 73:
        return 'C'
    elif average >= 60 and average < 68:
        return 'D'
    else:
        return 'F'
    
```

```python
print('{} -> {}'.format(83, get_letter_grade(83)))
print('{} -> {}'.format(40, get_letter_grade(40)))
print('{} -> {}'.format(100, get_letter_grade(100)))
```

```python
new_df['Final Average'].apply(get_letter_grade)
```

```python
new_df['Final Average'].apply(get_letter_grade) == new_df['Final Letter']
```

* As you can see, using .apply() or individual .loc[] assignments are both the same.
* However, you can also use assignment to do the apply:

```python
new_df['Final Letter'] = new_df['Final Average'].apply(get_letter_grade)
```
