```python
more_students_df['Final Average'] = (more_students_df['Midterm'] * .20) + (more_students_df['Final'] * .30) + (more_students_df['Attendance'] * 10 * .25) + (more_students_df['HW'] * 10 * .20)
```

```python
def get_letter_grade(average):
    if average >= 89:
        return 'A'
    elif average >= 83 and average < 89:
        return 'B+'
    elif average >= 79 and average < 83:
        return 'B'
    elif average >= 73 and average < 79:
        return 'C+'
    elif average >= 68 and average < 73:
        return 'C'
    elif average >= 60 and average < 68:
        return 'D'
    else:
        return 'F'
    
```

```python
def get_numeric_bin(average):
    return (average // 10) * 10
```

```python
more_students_df['Final Letter'] = more_students_df['Final Average'].apply(get_letter_grade)
```

### Question 1* - Solution

Use a bar graph to display the median of the Midtem, Final, and Final Average columns, per major

### Additional Question:
Identify the data items, attributes, and the marks and channels used. Comment on the
overall performance of each major...Which major(s) appear to be doing the best? Why do you
think so?

#### Answer:
items: a particular student's score

attributes: midterm score, final score, final average

marks: bar (rectangle)

channel: vertical position (number of students), color (type of score), horizontal position (major)

The majors that appear to be doing the best aren't so clear, but you could argue BNFO and MATH because their average scores are higher than the rest.

```python
majors_groups_med = final_df.groupby('Majors').agg(np.median)
majors_groups_med.reset_index(inplace=True)
majors_groups_med
```

```python
label_locations = np.arange(majors_groups_med.shape[0])
labs = majors_groups_med['Majors'].values
w = 0.2
plt.bar(label_locations - w, majors_groups_med['Midterm'], width=w, label='Midterm', color='red')
plt.bar(label_locations, majors_groups_med['Final'], width=w, label='Final', color='green')
plt.bar(label_locations + w, majors_groups_med['Final Average'], width=w, label='Final Avg', color='blue')
plt.gcf().set_size_inches(12, 6)
plt.title('Median Midterm, Final, and Final Average Scores')
plt.xlabel('Major')
plt.ylabel('Median Score')
plt.xticks(label_locations, labels=labs, rotation=45)
plt.legend()
plt.tight_layout() 
plt.show()
```
